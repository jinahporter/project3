```python
import pandas as pd
import os
```

```python
df = pd.read_csv('cleaned_subregion_data.csv')
df
```

```python
## getting all of the cities in TX 
tx_cities = df[(df["city"] == 'san antonio') | (df['city'] == 'houston') | (df['city'] =='dallas')]

tx_cities.head()

# check to see any null values 
tx_cities.isnull().any()
tx_cities.count()
```

```python
dallas_temp = pd.read_json('dallas.json')
dallas_temp.head()

# add column for city name before merging
dallas_temp['city'] = 'dallas'
dallas_temp.head()
```

```python
sa_temp = pd.read_json('san_antonio.json')
sa_temp.head()

# add column name
sa_temp['city'] = 'san antonio'

sa_temp.head()
```

```python
## combine all weather datasets

tx_weather = pd.concat([houston_temp, dallas_temp, sa_temp], axis = 0)
```
